```python
from sklearn.datasets import load_wine
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd
import numpy as np
from sklearn.cluster import AgglomerativeClustering,KMeans
from sklearn.metrics import silhouette_score
from sklearn.metrics.cluster import adjusted_rand_score
import warnings
warnings.filterwarnings("ignore")


wine = load_wine()

X,y = wine.data,wine.target
```

```python
labels = AgglomerativeClustering(n_clusters=3,linkage='single',metric='euclidean').fit_predict(X)
silhouette = silhouette_score(X,labels)
print(f"Distancia: euclidean, Linkage:single, Silhouette: {silhouette}")
```

```python
labels = AgglomerativeClustering(n_clusters=3,linkage='single',metric='cityblock').fit_predict(X)
silhouette = silhouette_score(X,labels)
print(f"Distancia: cityblock, Linkage:single, Silhouette: {silhouette}")
```

```python
labels = AgglomerativeClustering(n_clusters=3,linkage='average',metric='euclidean').fit_predict(X)
silhouette = silhouette_score(X,labels)
print(f"Distancia: euclidean, Linkage:average, Silhouette: {silhouette}")
```

```python
labels = AgglomerativeClustering(n_clusters=3,linkage='average',metric='cityblock').fit_predict(X)
silhouette = silhouette_score(X,labels)
print(f"Distancia: cityblock, Linkage:average, Silhouette: {silhouette}")
```

```python
labels = AgglomerativeClustering(n_clusters=3,linkage='complete',metric='euclidean').fit_predict(X)
silhouette = silhouette_score(X,labels)
print(f"Distancia: euclidean, Linkage:complete, Silhouette: {silhouette}")
```

```python
labels = AgglomerativeClustering(n_clusters=3,linkage='complete',metric='cityblock').fit_predict(X)
silhouette = silhouette_score(X,labels)
print(f"Distancia: cityblock, Linkage:complete, Silhouette: {silhouette}")
```

```python
kmeans = KMeans(n_clusters =3,n_init=10,init='random',max_iter=300,random_state=0)
kmeans.fit_predict(X)
```

```python
labels = kmeans.labels_
silhouette = silhouette_score(X,labels)
print(f"KMeans silhouette eh {silhouette}")
```

```python
labels = AgglomerativeClustering(n_clusters=3,linkage='average',metric='euclidean').fit_predict(X)
adjusted_rand = adjusted_rand_score(y,labels)
print(f"Indice rand do melhor eh {adjusted_rand}")
```

Melhor config: Aglomerativo com linkage average e distancia euclidiana, com silhueta média de 0.6100753288756406, rand index de 0.292626917173625

```python
dataset = [['Milk', 'Onion', 'Nutmeg', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Eggs'],
           ['Milk', 'Unicorn', 'Corn', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Ice cream', 'Eggs']]
```

```python
dataset
```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
frequent_itemsets = apriori(df, min_support=0.5, use_colnames=True)
frequent_itemsets
```

```python
df_ar = association_rules(frequent_itemsets, metric = "confidence", min_threshold = 0.7)
df_ar
```

2 regras geradas, (Onion)->(Eggs) e (Eggs)->(Onion), (Onion) é a melhor já que 100% das vezes que ela aparece, eggs também aparece
